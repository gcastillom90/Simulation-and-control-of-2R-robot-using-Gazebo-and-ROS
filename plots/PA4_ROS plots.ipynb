```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

j1 = open("joint1_data6.txt","r")		      #Create text files to save the data of joint angles
j2 = open("joint2_data6.txt","r")
if j1.mode == "r":
    joint1 = j1.read()
    joint1 = joint1.split('\n')
    del joint1[-1]
    joint1 = list(map(float, joint1))


if j2.mode == "r":
    joint2 = j2.read()
    joint2 = joint2.split('\n')
    del joint2[-1]
    joint2 = list(map(float, joint2))

plt.plot(joint1[0:2000],color="blue", linewidth=1.5, linestyle="-", label="$\Theta_1$")   
plt.plot(joint2[0:2000],color="red", linewidth=1.5, linestyle="-", label="$\Theta_2$")

#Plot legend
plt.legend(loc='upper right')

#Axis format
ax = plt.gca()  # gca stands for 'get current axis'


#Axis labels
ax.set_xlabel('Time [msx10]')
ax.set_ylabel('Joint angle [rad]')

plt.show()
```
